```python
import pandas as pd
import requests
from dotenv import load_dotenv
import os
```

```python
load_dotenv()
API_KEY = os.getenv("API_KEY")
```

## taking movie data from TMDB API

```python
combined_df = pd.DataFrame()

imageUrl = "https://image.tmdb.org/t/p/original"

headers = {
    "accept": "application/json",
    "Authorization": "Bearer " + API_KEY}

for i in range(1, 54):
    page = str(i)
    url = "https://api.themoviedb.org/3/discover/movie?include_adult=false&include_video=false&language=en-US&page="+ page +"&sort_by=popularity.desc"

    response = requests.get(url, headers=headers).json()

    # Create a DataFrame for the current page
    df = pd.DataFrame.from_dict(response['results'])
    df['poster_path'] = imageUrl + df['poster_path'].astype(str)

    # Append the data to the combined DataFrame
    combined_df = pd.concat([combined_df, df], ignore_index=True)

# Save the combined DataFrame to the CSV file
combined_df.to_csv('movieswithposters.csv', encoding='utf-8' ,index=False)
```

## taking series data from TMDB API

```python
combined_df2 = pd.DataFrame()

for i in range(1, 54):
    page = str(i)
    url = "https://api.themoviedb.org/3/discover/tv?include_adult=false&include_null_first_air_dates=false&language=en-US&page="+ page +"&sort_by=popularity.desc"

    response = requests.get(url, headers=headers).json()

    # Create a DataFrame for the current page
    df = pd.DataFrame.from_dict(response['results'])
    df['poster_path'] = imageUrl + df['poster_path'].astype(str)

    # Append the data to the combined DataFrame
    combined_df2 = pd.concat([combined_df2, df], ignore_index=True)

# Save the combined DataFrame to the CSV file
combined_df2.to_csv('serieswithposters.csv', encoding='utf-8' ,index=False)
```

## taking movie genres from TMDB API

```python
url = "https://api.themoviedb.org/3/genre/movie/list?language=en"
response = requests.get(url, headers=headers).json()
df = pd.DataFrame.from_dict(response['genres'])
df.to_csv('movieGenres.csv', encoding='utf-8' ,index=False)
```

## taking series genre from TMDB API

```python
url2 = "https://api.themoviedb.org/3/genre/tv/list?language=en"
response2 = requests.get(url2, headers=headers).json()
df2 = pd.DataFrame.from_dict(response2['genres'])
df2.to_csv('seriesGenres.csv',encoding='utf-8',index=False)
```
