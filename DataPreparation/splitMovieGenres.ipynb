```python
def string_to_float_array_method(input_string):
    return [int(x) for x in input_string.split()]

# Assuming dfMovie is your DataFrame
for i in range(len(dfMovie['genre_ids'])):
    dfMovie['genre_ids'][i] = string_to_float_array_method(dfMovie['genre_ids'][i])
```

```python
def find_max_length(lst):
    if isinstance(lst, list):
        return len(lst)
    else:
        return 0
```

```python
def max_length_in_column(column):
    lengths = column.apply(find_max_length)
    max_length = lengths.max()
    return max_length
```

```python
max_length = max_length_in_column(dfMovie['genre_ids'])
max_length
```

```python
new_columns = [f'Genre{i+1}' for i in range(max_length)]

dfMovie[new_columns] = np.nan
```
