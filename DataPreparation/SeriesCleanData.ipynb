```python
series['Genre1'] = pd.to_numeric(series['Genre1'], errors='coerce', downcast='integer')
series['Genre1'] = series['Genre1'].fillna(0).astype('int64')

series['Genre2'] = pd.to_numeric(series['Genre2'], errors='coerce', downcast='integer')
series['Genre2'] = series['Genre2'].fillna(0).astype('int64')
```

```python
series = series.drop('backdrop_path',axis=1)
series = series.drop('genre_ids',axis=1)
series = series.drop('original_name',axis=1)
series = series.drop('popularity',axis=1)
series = series.drop('vote_count',axis=1)
```
